```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#On génère d'abords des variables aléatoires uniformément distribuées avec Numpy

vecteur=np.random.rand(30)

plot=plt.scatter(range(30),vecteur,linewidths=3,color="green")
```

```python
#Essayant maintenant avec une matrice 

mat_alea_a_imprimer=np.random.rand(2,3)
print(mat_alea_a_imprimer)

mat_alea_a_grapher=np.random.rand(2,30)

plot=plt.scatter(range(30),mat_alea_a_grapher[0,:])
plot=plt.scatter(range(30),mat_alea_a_grapher[1,:])
```

```python
#Quelques notions de base de la librairie Python

#Considerons 100 variables aléatoire et on tire d'eux un échantillon de 200 individus

matrice_aleatoire=np.random.rand(100,200)

#la somme de Numpy avec un 0 donne la somme des lignes
#avec un 1 donne la somme des colonnes
#Sinon donne la somme de toute la matrice 
somme=np.sum(matrice_aleatoire,0)
print("la taille de la variable somme est {}".format(somme.shape))
```

```python
#On essaye de visualiser cette somme

plot=plt.scatter(range(200),somme)
```

```python
#l'histogramme est beacoup mieux pour visualiser une distribution
plot=plt.hist(somme)
```

```python
#Et si on essaye avec un nombre plus grand que 200

matrice_aleatoire=np.random.rand(10000,20000)

somme=np.sum(matrice_aleatoire,0)
plot=plt.hist(somme,bins=100)
#On remarque bin que notre variable suit la loi gaussienne(normale)
```
